```python
# category = input() #enter category here , for example 'IT Assets' 'Non IT Assets'
category = "IT Assets" #comment this out when using
filter = [True if x==category else False for x in data["Category"]]
category_split = data.iloc[filter,:]

X = Countries = sorted(category_split["Country"].unique()) # Xlabel for graph

country_wise = category_split.groupby(['Country']).sum()

# ENTER DATE RANGE OVER HERE
range = "Apr-20 to Mar-21"
Y = [x for x in country_wise[range]] # Ylabel for graph

fig = plt.figure(figsize = (10, 5))

# creating the bar plot
plt.bar(X,Y, color ='maroon',width = 0.4)
plt.xlabel("Countries")
plt.ylabel("Expenditure (in USD)")
plt.title(f"Expenditure on {category} during year {range}")
plt.show()
```
